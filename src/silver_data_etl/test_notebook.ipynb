```python
import re

re.findall(
    r"[NBRQK]?[a-h]?[1-8]?x?[a-h][1-8][\+|#]?|O-O(?:-O)?",
    chess_string,
)
```

```python
chess_objects = []
chess_object = {}

for line in chess_string.splitlines():
    if line.startswith("["):
        key, value = line.strip(" []").split(" ", 1)
        value = value.strip('"')
        chess_object[key] = value
    else:
        chess_object.setdefault("moves", "")
        chess_object["moves"] += line.strip() + "\n"
    if line.endswith("1-0") or line.endswith("0-1") or line.endswith("1/2-1/2"):

        all_moves = []
        all_times = []

        move_matches = re.findall(
            r" ([NBRQK]?[a-h]?[1-8]?x?[a-h][1-8][NBRQ]?[\+|#]?|O-O(-O)?) ",
            chess_object["moves"],
        )

        for move in move_matches:
            all_moves.append(move[0])

        time_matches = re.findall(r"%clk ([^\]]+)", chess_object["moves"])

        for time in time_matches:
            all_times.append(time)

        chess_object["moves"] = all_moves
        chess_object["times"] = all_times

        chess_objects.append(chess_object)
        chess_object = {}
```
